```python
import pandas as pd

# Load your data
data = pd.read_excel(r"C:\Users\nisha\Downloads\datafile.xls")

# Convert all column names to strings
data.columns = data.columns.astype(str)

# Now your original code will work perfectly!
top_10_states = data.sort_values(by='2017', ascending=False).head(10)

print(top_10_states)
```

```python
#Bar Chart: Top 10 States by Cyber Crimes in 2017

import matplotlib.pyplot as plt

# --- Step 1: Prepare the data ---
# Sort the DataFrame by the '2018' column in descending order and get the top 10
top_10_states = data.sort_values(by='2017', ascending=False).head(10)

# --- Step 2: Create the plot ---
plt.figure(figsize=(12, 8))  # Create a figure with a custom size for better readability

# Create a horizontal bar chart (barh)
plt.barh(top_10_states['State/UT'], top_10_states['2017'], color='skyblue')

# --- Step 3: Add labels and title ---
plt.xlabel('Number of Cyber Crimes')
plt.title('Top 10 States by Number of Cyber Crimes in 2017')
plt.gca().invert_yaxis()  # Invert y-axis to show the highest value at the top

# --- Step 4: Display the plot ---
plt.tight_layout()  # Adjust layout to make sure everything fits
plt.show()
```

```python
#Scatter Plot: Population vs. Rate of Crime

import matplotlib.pyplot as plt

# --- Step 1: Create the plot ---
plt.figure(figsize=(10, 6))

# Use column names exactly as they appear in your DataFrame
plt.scatter(data['Mid-Year Projected Population (in Lakhs) (2018)+'], 
            data['Rate of Total Cyber Crimes (2018)++'], 
            alpha=0.6) # alpha makes points slightly transparent to see overlaps

# --- Step 2: Add labels and title ---
plt.xlabel('Projected Population in 2017 (in Lakhs)')
plt.ylabel('Rate of Total Cyber Crimes (2017)')
plt.title('Population vs. Rate of Cyber Crime in Indian States/UTs (2017)')
plt.grid(True) # Add a grid for easier reading

# --- Step 3: Display the plot ---
plt.show()
```
